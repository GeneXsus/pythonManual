## 1 - Despliegue en PythonAnywhere

[PythonAnywhere](https://www.pythonanywhere.com/) es una plataforma en la nube que permite a los desarrolladores escribir, ejecutar y alojar aplicaciones Python directamente desde su navegador, sin necesidad de configurar un servidor físico o virtual. Es especialmente popular para el alojamiento de aplicaciones web construidas en frameworks como Flask y Django, y es muy útil para desarrolladores que desean implementar aplicaciones rápidamente sin lidiar con la infraestructura del servidor.

<br>

**Características principales de PythonAnywhere**

1. Alojamiento web para aplicaciones Python. Podemos desplegar aplicaciones web escritas en Python, como las que usan Flask o Django con solo unos pocos clics. PythonAnywhere configura y gestiona el servidor web, lo que facilita el proceso de despliegue.


2. Acceso a un editor de código en línea. PythonAnywhere ofrece un editor de código en el navegador, así que podemos escribir y editar nuestro código sin necesidad de un entorno de desarrollo local.

3. Bases de datos integradas. Soporta bases de datos como MySQL y PostgreSQL, que puedes utilizar para tu aplicación web directamente en la plataforma.

4. Terminal bash en el navegador. PythonAnywhere proporciona una terminal de Linux accesible desde el navegador, lo que nos permite ejecutar comandos Bash, instalar paquetes y realizar otras tareas de administración.

5. Ejecución de tareas programadas (Scheduled Tasks). Podemos configurar tareas programadas, lo cual es útil para ejecutar scripts periódicos como análisis de datos, envíos de correos o tareas de mantenimiento.

6. Entorno de pruebas y producción. Permite probar la aplicación en la plataforma antes de lanzarla a producción, lo cual facilita el flujo de desarrollo y despliegue.

<br>

**Ventajas de PythonAnywhere**

+ Facilidad de uso: No necesitamos conocimientos avanzados de administración de servidores; PythonAnywhere se encarga de la infraestructura.

+ Gratuito y de pago: Ofrece un plan gratuito que permite experimentar y probar la plataforma, además de planes de pago con más capacidad y funciones.

+ Alojamiento de aplicaciones web rápido: Ideal para desarrolladores que quieren centrarse en el desarrollo de la aplicación y evitar la configuración de servidores.

<br>

**¿Para qué se usa PythonAnywhere?**

+ Despliegue rápido de aplicaciones web en proyectos pequeños y medianos.

+ Probar y experimentar con scripts de Python sin necesidad de una configuración local.

+ Alojar proyectos de aprendizaje y portafolio para mostrar aplicaciones en línea.

+ Automatizar tareas programadas con scripts de Python.
