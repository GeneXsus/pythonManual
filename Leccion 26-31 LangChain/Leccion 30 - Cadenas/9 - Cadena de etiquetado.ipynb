## 1 - Cadena de etiquetado

En LangChain, `create_tagging_chain` es una función utilizada para crear una cadena de etiquetado, que sirve para clasificar o asignar etiquetas a texto basado en un esquema definido. El objetivo principal de esta función es automatizar el proceso de análisis de texto y etiquetado utilizando un modelo de lenguaje como motor. Esto es útil en tareas de NLP, donde se necesita categorizar texto en diferentes clases o categorías.

```python
# cargamos la API KEY de OpenAI

from dotenv import load_dotenv 
import os

# carga de variables de entorno
load_dotenv()


# api key openai, nombre que tiene por defecto en LangChain
OPENAI_API_KEY = os.getenv('OPENAI_API_KEY')
```

```python
# importamos librerias

from langchain_openai import ChatOpenAI

from langchain.prompts import ChatPromptTemplate

from langchain.schema.output_parser import StrOutputParser

from langchain.utilities import SQLDatabase

from langchain.schema.runnable import RunnablePassthrough

from langchain.chains import create_tagging_chain
```

```python
# iniciamos el modelo de OpenAI

modelo = ChatOpenAI()
```

```python
# parser de salida, transforma la salida a string

parser = StrOutputParser()
```

```python
# esquema de etiquetas

esquema = {'properties': {'sentiment': {'type': 'string'},
                          'aggressiveness': {'type': 'integer'},
                          'language': {'type': 'string'}
                         }
          }
```

```python
# creacion de la cadena

cadena = create_tagging_chain(llm=modelo, schema=esquema)
```

```python
# respuesta cadena

texto = 'Estoy increiblemente contento de haberte conocido! Creo que seremos muy buenos amigos!'

cadena.invoke(texto)
```

```python
# respuesta cadena

texto = 'Estoy muy enojado contigo! Te voy a dar tu merecido!'

cadena.invoke(texto)
```
