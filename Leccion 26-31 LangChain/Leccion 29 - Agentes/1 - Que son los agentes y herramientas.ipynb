## 1 - ¿Qué es un agente?

Un agente en LangChain es un componente avanzado que toma decisiones dinámicas sobre qué acciones realizar utilizando un modelo de lenguaje. A diferencia de las cadenas tradicionales, donde las acciones están predefinidas y codificadas de forma rígida, los agentes en LangChain son más flexibles y adaptativos. Utilizan el razonamiento proporcionado por el modelo de lenguaje para decidir qué pasos tomar, qué herramientas usar y en qué orden hacerlo, dependiendo del contexto y del objetivo de la tarea.

<br>

**Componentes de un agente**:

1. Modelo de lenguaje: Es el núcleo del agente, utilizado para tomar decisiones sobre los pasos a seguir.

2. Entrada del usuario: El objetivo o consulta que proporciona el usuario.

3. Herramientas: Interfaces que el agente puede usar para realizar acciones específicas, como buscar en la web, acceder a APIs, ejecutar comandos, etc.

4. Historial de pasos intermedios: Un registro de las acciones previas que el agente ha tomado y las salidas resultantes de esas acciones.

<br>

**Tipos de salidas de un agente**:

+ AgentActions: Las siguientes acciones a realizar, que incluyen la herramienta a utilizar y la entrada correspondiente.

+ AgentFinish: La respuesta final del agente, cuando ya no es necesario realizar más acciones.

<br>

Por ejemplo, un agente podría recibir una consulta, decidir que necesita hacer una búsqueda en la web, y luego, con base en los resultados de esa búsqueda, decidir la siguiente acción o responder al usuario. Esto hace que los agentes sean especialmente útiles para tareas complejas o interactivas donde se necesita razonamiento dinámico.


<br>



**Tipos de Agentes**

LangChain ofrece varios tipos de agentes, cada uno adecuado para casos de uso específicos. Algunos de los agentes disponibles son los siguientes:

+ Zero-shot ReAct: Este agente utiliza el marco ReAct para elegir herramientas basándose únicamente en sus descripciones. Requiere descripciones para cada herramienta y es muy versátil.

+ Structured input ReAct: Este agente maneja herramientas con múltiples entradas y es adecuado para tareas complejas, como navegar en un navegador web. Utiliza el esquema de argumentos de las herramientas para entradas estructuradas.

+ OpenAI Functions: Diseñado específicamente para modelos ajustados para llamadas a funciones, este agente es compatible con modelos como gpt-3.5-turbo y gpt-4. 

+ Conversacional: Diseñado para entornos conversacionales, este agente utiliza ReAct para la selección de herramientas y usa memoria para recordar interacciones previas.

+ Self-ask with search: Este agente se basa en una sola herramienta, "Intermediate Answer", que busca respuestas factuales a preguntas. 

+ ReAct document store: Este agente interactúa con un almacén de documentos usando el marco ReAct. Requiere herramientas de "Búsqueda" y "Consulta" (Search y Lookup).

## 2 - Herramientas

Las herramientas son interfaces que un agente puede utilizar para interactuar con el mundo. Permiten a los agentes realizar diversas tareas, como buscar en la web, ejecutar comandos en la terminal o acceder a APIs externas. En LangChain, las herramientas son esenciales para extender las capacidades de los agentes y permitirles llevar a cabo tareas diversas. Existen una multitud de [herramientas](https://python.langchain.com/docs/integrations/tools/) en LangChain. Veamos algunos ejemplos.


### 2.1 - DuckDuckGoSearchRun


Podemos utilizar `DuckDuckGoSearchRun` para realizar búsquedas en la web. Para usar este objeto necesitamos ejecutar el siguiente comando:

```bash
pip install duckduckgo-search
```

````python
from langchain.tools import DuckDuckGoSearchRun
````

````python
buscador = DuckDuckGoSearchRun()
````

````python
buscador.run('que es langchain')
````

Para obtener más información adicional, por ejemplo el enlace o la fuente, podemos utilizar `DuckDuckGoSearchResults`.

````python
from langchain_community.tools import DuckDuckGoSearchResults
````

````python
buscador = DuckDuckGoSearchResults()
````

````python
buscador.run('que es langchain')
````

### 2.2 - ShellTool

El objeto `ShellTool` nos permite ejecutar comandos de terminal como una herramienta de LangChain.

````python
from langchain.tools import ShellTool
````

````python
shell = ShellTool()
````

````python
resultado = shell.run({'commands': ['ls', 'echo hola']})
````

````python
resultado.split('\n')
````

### 2.3 - YouTubeSearchTool

El paquete `YouTubeSearchTool` busca videos en YouTube evitando el uso de su API, que tiene muchas limitaciones de tasa. Utiliza el formulario en la página de inicio de YouTube y extrae la información de la página de resultados.

Para usar esta herramienta debemos ejecutar el siguiente comando:

```bash
pip install youtube_search
```

````python
from langchain.tools import YouTubeSearchTool
````

````python
youtube = YouTubeSearchTool()
````

````python
busqueda = 'Soil & Pimp Sessions'
````

````python
youtube.run(busqueda)
````

Por defecto, se devuelven 2 links. Podemos añadir a la busqueda el numero de links que queremos.

````python
busqueda = 'Soil & Pimp Sessions, 5'
````

````python
youtube.run(busqueda)
````

### 2.5 - Herramienta custom

Podemos definir herramientas propias para el uso de los agentes. Para ello, creamos una clase que hereda del objeto `BaseTool` de LangChain. Vamos a crear una clase para calcular la longitud de una circunferencia dado el radio. Esta clase necesita los atributos `name` y `description` además de los métodos `_run` y `_arun` para poder ejecutarse.

````python
from langchain.tools import BaseTool

from math import pi
from typing import Union
````

````python
class Circunferencia(BaseTool):
    
    # atributos, se necesitan estos nombres de atributos y que esten tipados
    name: str = 'Calcula longitud de la circunferencia'
    
    description: str = 'Usa esta herramienta para calcular la circunferencia dado el radio'
        
    
    def _run(self, radio: Union[int, float]):
        """
        Método para calcular la longitud
        Necesita el nombre `_run` para poder ser ejecutada
        
        Params:
        radio: int o float, radio de la circunferencia
        
        Return:
        float con la longitud de la circunferencia
        """
        return  2.0 * pi * float(radio)
    
    
    
    def _arun(self, radio: Union[int, float]):
        """
        Método para ejecución asincrona.
        No implementado.
        """
        raise NotImplementedError('Esta herramienta no soporta async')
````

````python
# uso herramienta custom

Circunferencia().run(tool_input={'radio': 0.5})
````

### 2.6 - Lista de herramientas

Podemos crear una lista de herramientas para que el agente pueda tener acceso a varias herramientas a la vez. Esto se hace con el objeto `Tool`. Vamos a hacer una lista con algunas de las herramientas anteriores.

````python
from langchain.agents import Tool
````

````python
# añadimos DuckDuckGo

herramientas = [Tool(name='DuckDuckGo',
                     func=buscador.run,
                     description='Utilidad de busqueda en internet')]
````

````python
type(herramientas)
````

````python
# añadimos wikipedia

herramientas.append(Tool(name='Wikipedia',
                         func=wikipedia.run,
                         description='Utilidad de busqueda en wikipedia'))
````

````python
# añadimos youtube

herramientas.append(Tool(name='Youtube',
                         func=youtube.run,
                         description='Utilidad de busqueda en youtube'))
````
