## 1 - ¿Qué es AWS S3?

Amazon Simple Storage Service (Amazon S3) es un servicio de almacenamiento en la nube proporcionado por Amazon Web Services (AWS). Es ampliamente utilizado para almacenar y recuperar cualquier cantidad de datos desde cualquier ubicación, ya sea en aplicaciones web, móviles, de IoT, o para respaldo y recuperación.

<br>

**Características clave de Amazon S3**

1. Almacenamiento escalable: S3 permite almacenar cantidades masivas de datos, desde unos pocos bytes hasta petabytes. Podemos comenzar pequeño y escalar según nuestras necesidades.


2. Alta durabilidad y disponibilidad: Amazon S3 ofrece una durabilidad del 99.999999999%, lo que significa que los datos están protegidos contra fallos de hardware. Tiene una alta disponibilidad con opciones de replicación en múltiples regiones.


3. Flexible: Podemos almacenar objetos de cualquier tipo, como documentos, imágenes, videos, backups o logs. Cada objeto puede tener hasta 5 TB.


4. Acceso y control granular: Podemos definir políticas detalladas para controlar quién accede a qué datos y bajo qué condiciones. Compatible con cifrado en reposo y en tránsito para proteger la seguridad de los datos.


5. Opciones de almacenamiento: S3 tiene diferentes clases de almacenamiento según nuestras necesidades:
    + S3 Standard: Ideal para datos a los que accedes frecuentemente.
    + S3 Intelligent-Tiering: Optimiza automáticamente los costos moviendo datos entre clases de almacenamiento según su uso.
    + S3 Glacier y S3 Glacier Deep Archive: Para almacenamiento de datos a largo plazo y recuperación esporádica.


6. Integración con otros servicios de AWS: Se integra con servicios como AWS Lambda, Amazon CloudFront, AWS Glue y otros, permitiendo flujos de trabajo y análisis de datos.


<br>

**Casos de uso comunes**

1. Almacenamiento de contenido estático: Imágenes, videos, archivos CSS/JS y otros recursos web.


2. Backup y recuperación: Respaldo seguro de datos y restauración ante desastres.


3. Big Data y análisis: Almacenamiento de grandes volúmenes de datos para análisis.


4. Distribución de contenido: Se combina con servicios como Amazon CloudFront para entregar contenido de baja latencia.


5. Archivos y datos de larga duración: Almacenamiento de datos de archivo como registros financieros o históricos.

<br>

**Cómo funciona**

1. Buckets: Los datos en S3 se almacenan en contenedores llamados buckets, que son los repositorios de almacenamiento.


2. Objetos: Los datos se almacenan como objetos, que consisten en datos, metadatos y un identificador único.


3. Regiones: Puedes seleccionar una región para almacenar tus datos, optimizando costos y latencia.


4. Modelo de pago: S3 sigue un modelo de pago por uso. Pagas por el almacenamiento utilizado, el tráfico de salida y las solicitudes realizadas.

## 3 - CRUD

Recordemos que CRUD representa las cuatro operaciones básicas que se pueden realizar en cualquier sistema de almacenamiento de datos. Estas operaciones son:

+ **Create (Crear)**: Agregar nuevos datos al sistema.

+ **Read (Leer)**: Recuperar o consultar datos existentes.

+ **Update (Actualizar)**: Modificar datos existentes.

+ **Delete (Eliminar)**: Borrar datos del sistema.

````python
# primero importamos la secret key y la key id

import os                           # libreria del sistema operativo
from dotenv import load_dotenv      # carga variables de entorno 


load_dotenv()

AWS_SECRET_KEY_S3 = os.getenv('AWS_SECRET_KEY_S3')
AWS_KEY_ID = os.getenv('AWS_KEY_ID')
````

````python
# importamos boto

import boto3
from botocore.client import Config
````

````python
# iniciamos el cliente

cliente = boto3.client('s3',
                       aws_access_key_id=AWS_KEY_ID,
                       aws_secret_access_key=AWS_SECRET_KEY_S3,
                       region_name='eu-central-1',
                       endpoint_url='https://s3.eu-central-1.amazonaws.com',
                       config=Config(signature_version='s3v4'))
````

````python
# nombre del bucket

BUCKET = 'yonaejemplo'
````

### 3.1 - Crear y actualizar

Crear y actualizar archivos en un bucket es básicamente lo mismo, los archivos si no existen se crean y si existen se sobreescriben. Veamos cómo hacerlo:

````python
cliente.upload_file(Bucket=BUCKET,            # nombre del bucket
                    Key='carpeta/prueba.txt', # ruta dentro del bucket 
                    Filename='prueba.txt'     # ruta al archivo local 
                   )
````

Podemos comprobar la subida del archivo en la consola del bucket en AWS.

### 3.2 - Leer y descargar

````python
# lectura directa de archivo

s3 = boto3.resource('s3',
                    region_name='eu-central-1', 
                    aws_access_key_id=AWS_KEY_ID, 
                    aws_secret_access_key=AWS_SECRET_KEY_S3)

archivo = s3.Object(bucket_name=BUCKET,
                    key='carpeta/prueba.txt')


archivo.get()['Body'].read()
````

````python
# descarga

cliente.download_file(Bucket=BUCKET,            # nombre del bucket
                    Key='carpeta/prueba.txt',   # ruta dentro del bucket 
                    Filename='prueba.txt'       # ruta al archivo local 
                    )
````

### 3.3 - Borrar

````python
archivos = [e['Key'] for e in cliente.list_objects(Bucket=BUCKET, 
                                                   Prefix='carpeta/prueba.txt')['Contents']]
        
archivos
````

````python
for e in archivos:
    
    cliente.delete_object(Bucket=BUCKET, Key=e)
````

````python
# comprobación

try:
    s3 = boto3.resource('s3',
                        region_name='eu-central-1', 
                        aws_access_key_id=AWS_KEY_ID, 
                        aws_secret_access_key=AWS_SECRET_KEY_S3)

    archivo = s3.Object(bucket_name=BUCKET,
                        key='carpeta/prueba.txt')


    archivo.get()['Body'].read()
    
except Exception as e:
    print(e)
````

## 4 - Resumen

Vamos a crear una clase con todos los métodos para CRUD:

````python
class S3_AWS:

    def __init__(self, bucket):

        self.bucket = bucket
        self.region = 'eu-central-1'
        self.AWS_SECRET_KEY_S3 = AWS_SECRET_KEY_S3
        self.AWS_KEY_ID = AWS_KEY_ID

        self.cliente = boto3.client('s3',
                                    aws_access_key_id=self.AWS_KEY_ID,
                                    aws_secret_access_key=self.AWS_SECRET_KEY_S3,
                                    region_name=self.region,
                                    endpoint_url=f'https://s3.{self.region}.amazonaws.com',
                                    config=Config(signature_version='s3v4'))
        
        
        
    def subir_archivo(self, ruta_s3: str, ruta_local: str) -> None:
        
        print('Subiendo archivo...')
        
        self.cliente.upload_file(Bucket=self.bucket, 
                                 Key=ruta_s3,                     
                                 Filename=ruta_local)
        
        
        print('Archivo subido.')
        
    
    
    def leer_archivo(self, ruta_s3: str) -> str:
        
        s3 = boto3.resource('s3',
                            region_name=self.region, 
                            aws_access_key_id=self.AWS_KEY_ID, 
                            aws_secret_access_key=self.AWS_SECRET_KEY_S3)

        archivo = s3.Object(bucket_name=self.bucket,
                            key=ruta_s3)


        return archivo.get()['Body'].read()
        
      
    
    
    def descargar_archivo(self, ruta_s3: str, ruta_local: str) -> None:
        
        print('Descargando archivo...')
        
        self.cliente.download_file(Bucket=self.bucket, 
                                   Key=ruta_s3,                     
                                   Filename=ruta_local)
        
        
        print('Archivo descargado.')
        
        
    def borrar_archivo(self, ruta_s3: str):
        
        print('Borrando archivo...')
        
        archivos = [e['Key'] for e in cliente.list_objects(Bucket=self.bucket, 
                                                           Prefix=ruta_s3)['Contents']]
        
        for e in archivos:
            cliente.delete_object(Bucket=self.bucket, Key=e)
            
        print('Archivo borrado.')
````

````python
ruta_local = 'prueba.txt'

ruta_s3 = 'carpeta/prueba.txt'
````

````python
s3.leer_archivo(ruta_s3)
````

````python
s3.descargar_archivo(ruta_s3, ruta_local)
````
