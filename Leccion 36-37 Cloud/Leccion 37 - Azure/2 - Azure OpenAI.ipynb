Existen varios servicios de inteligencia artificial en Azure, desde reconocimiento facial y visión por computador hasta reconocimiento de documentos con OCR, reconocimiento óptico de caracteres por las siglas en inglés. Nosotros vamos a usar `Azure OpenAI` para crear un chatbot desde Azure en vez de usarlo desde la API de OpenAI. El costo es básicamente el mismo, pero la velocidad de la respuesta del chat es mucho mayor. Para proyectos en producción es recomendable usar directamente este servicio.

Una vez creada la nueva instancia del modelo, en su descripción, necesitamos guardar dos datos para poder tener el punto de conexión externo:

+ `URI de destino`: este es el endpoint a cual tendremos que llamar, debemos asegurarnos que la version de la API está incluida en esta string (?api-version=2024-08-01-preview por ejemplo)

+ `Clave`: esta es la API Key que necesitamos para registrarnos.

Estos datos se deberían de guardar en el archivo `.env` de nuestro entorno virtual y cargarla con dotenv por seguridad.

## 2 - Uso del ChatBot

En primer lugar vamos a cargar las credenciales necesarias desde el archivo `.env`:

```python
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
# endpoint y api key

AZURE_OPENAI_ENDPOINT = os.getenv('AZURE_OPENAI_ENDPOINT')

AZURE_OPENAI_API_KEY = os.getenv('AZURE_OPENAI_API_KEY')
```

Una vez importadas las credenciales, cargamos el objeto `AzureChatOpenAI` con LangChain para poder usar el chatbot que hemos creado:

```python
from langchain_openai import AzureChatOpenAI
```

```python
llm = AzureChatOpenAI(deployment_name='gpt-4o', temperature=0.7)
```

```python
respuesta = llm.invoke('Dime los principales emperadores romanos.').content
```

```python
print(respuesta)
```

Ahora que tenemos el llm funcionando, podemos incorporarlo a una cadena de LangChain como ya hemos visto en módulos anteriores y usarlo de la misma manera que usaríamos el chat desde OpenAI.
